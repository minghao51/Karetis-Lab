```python
Insee = pd.read_csv('Data/L1_TV_InseeData.csv', index_col='TV').drop(["Unnamed: 0","Population"], axis=1)
TownData = pd.read_csv('Data/L1_TV_TownData.csv', index_col='TV').drop("Unnamed: 0", axis=1)
```

```python
Data = Insee.join(TownData)
```

```python
dt= pd.DataFrame()
dt['Population_17_pct'] = Data['Population_17YO']/Data['Population']
dt['Population18_pct'] = Data['Population18_39YO']/Data['Population']
dt['Population40_pct'] = Data['Population40_64YO']/Data['Population']
dt['Population65_pct'] = Data['Population65_79YO']/Data['Population']
dt['Population80_pct'] = Data['Population80pYO']/Data['Population']

cur_list = [i for i in Data.columns if '_pct' in i]
for i in cur_list:
    # print(i)
    dt[i] = Data[i]
```
